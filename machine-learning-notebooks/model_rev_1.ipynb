# Preprocessing and feature engineering


```python
def copy_df(df:pd.DataFrame):
    return df.copy()
```

```python
def mode_to_numerical(df: pd.DataFrame):
    df.loc[:, "mode"] = (df["mode"]=="operation").astype(float)
    return df
```

```python
def drop_na(df: pd.DataFrame):
    return df.dropna(axis=0)
```

```python
train_main_features_prep = (
    train_main_features.pipe(copy_df)
    .pipe(mode_to_numerical)
    .pipe(drop_na)
)
train_main_features_prep.head()
```

```python
# drop nas
def drop_target_nans(targets_df, train_raw_df):
    is_nan_row = train_raw_df.isna().any(axis=1)
    targets_df = targets_df[~is_nan_row]
    return targets_df
```

# build model

```python
@dataclass
class ModelResponse:
    model: Any
    train_mape: float
    test_mape: float
```

```python
def train_bolts_models(train_df, targets_df, model_name="linear"):
    if model_name == "linear":
        model = LinearRegression
    else:
        raise ValueError()
    
    res = {}
    for target_name, target_data in targets_df.iteritems():
        X_train, X_test, y_train, y_test = train_test_split(train_df, target_data, test_size = 0.8, shuffle=True)
        m = model()
        m.fit(X_train, y_train)
        y_pred_train = m.predict(X_train)
        y_pred_test = m.predict(X_test)
        train_mape = mean_absolute_percentage_error(y_train, y_pred_train)
        test_mape = mean_absolute_percentage_error(y_test, y_pred_test)
        res[target_name] = ModelResponse(model = m, train_mape=train_mape, test_mape=test_mape)
        
    return res
        
```

```python
model_responses = train_bolts_models(train_main_features_prep_norm, targets_df_prep)
model_responses
```
